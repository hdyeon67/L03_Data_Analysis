# [4] 데이터 확인하기

```python
import pandas as pd
```

```python
employee_data = {
    "Name": ["John Doe", "Jane Smith", "Peter Jones", "Linda Taylor", "Emily Clark", "David Smith", "Michael Brown", "Sarah Johnson"],
    "Position": ["Software Engineer", "Data Scientist", "Project Manager", "HR Specialist", "Data Scientist", "Software Engineer", "HR Specialist", "Project Manager"],
    "Office": ["New York", "San Francisco", "Berlin", "London", "San Francisco", "New York", "London", "Berlin"],
    "Age": [28, 34, 45, 32, 30, 38, 41, 29],
    "Salary": [95000, 125000, 75000, 85000, 115000, 105000, 90000, 80000],
    "Is_Manager":[True,False,False,True,True,False,True,False]
}
df = pd.DataFrame(employee_data)
```

## (4) 정렬하기

데이터 프레임은 (행, 열) -> 2차원 구조

그렇기 때문에 정렬하는데, 어떤 방향을 기준으로 정렬할지 알려줘야 한다.  
=> `sort_*` : 정렬하세요. `*`를 기준으로

1. 값을 기준으로 정렬하는 df.sort_values()
2. 행의 이름을 기준으로 정렬하는 df.sort_index()

```python
# 컬럼을 기준으로 정렬
# 기본값은 오름차순

df.sort_values(by='Salary') # 새로운 데이터 프레임 반환
```

```python
df # 기존 데이터 프레임은 바뀌지 않는다.
```

> 원본 데이터 프레임을 바꾸고 싶은 경우?

```python
df.sort_values(by='Salary',inplace=True)
```

```python
df
```

```python
# 내림차순 정렬
df.sort_values(by='Age', ascending=False)
```

```python
# 여러컬럼을 기준으로 정렬
df.sort_values(by=['Office','Salary'],ascending=[True,False])
```

(참고) 행의 이름을 기준으로 정렬하기

```python
df.sort_index(inplace=True)
df
```

> IQR = Q3 - Q1  
 => 3사분위수 - 1사분위수



```python
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
```

```python
IQR = Q3 - Q1
```
